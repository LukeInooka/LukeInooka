```python
for goal_deg in range(360):
    for current_deg in range(360):
        diff = abs(current_deg-goal_deg)
        if diff <= 180 :
            print(goal_deg,current_deg,diff)
        elif goal_deg > current_deg:
            print(goal_deg,current_deg,diff- 360)
        else:
            print(goal_deg,current_deg,360-diff)
```
